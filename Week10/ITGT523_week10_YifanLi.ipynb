```python
import cv2
import pyautogui
import numpy as np
```

```python
cap = cv2.VideoCapture(0)
PINK_MIN = np.array([130, 150, 150], np.uint8)
PINK_MAX = np.array([180, 220, 200], np.uint8)
centroid_x = 0
centroid_y = 0
# 640 x 480
# 3 parts: 640/3 ~ 200
x1 = 200
x2 = 400
# 4 parts: 480/4 = 120
y1 = 120
y2 = 240
```

```python
while(cap.isOpened()):
    ret, img = cap.read()
    img = cv2.flip(img, 1)

    cv2.namedWindow('Original', cv2.WINDOW_NORMAL)
    hsv = cv2.cvtColor(img, cv2.COLOR_BGR2HSV)
    frame_threshed = cv2.inRange(hsv, PINK_MIN, PINK_MAX)

    contours,hierarchy = cv2.findContours(frame_threshed, 1, 2)
    max_area = 0
    last_x = centroid_x
    last_y = centroid_y

    if contours:
        # Find the biggest region of the target
        for i in contours:
            area = cv2.contourArea(i)
            if area > max_area:
                max_area = area
                cnt = i

        x,y,w,h = cv2.boundingRect(cnt)
        cv2.rectangle(img,(x,y),(x+w,y+h),(180,255,133),2)
        centroid_x = (x + x+w)/2
        centroid_y = (y + y+h)/2

        # Draw the center
        cv2.circle(img, (int(centroid_x), int(centroid_y)), 2, (0,0,255), 2)
        cv2.line(img,(x1,0),(x1,y2),(255,0,0),5)
        cv2.line(img,(x2,0),(x2,y2),(255,0,0),5)
        cv2.line(img,(x1,y1),(x2,y1),(255,0,0),5)
        cv2.line(img,(x1,y2),(x2,y2),(255,0,0),5)

       # up-down move
        if centroid_x >= x1 and centroid_x <= x2:
            # up
            if centroid_y >= 0 and centroid_y <= y1:
                print('up')
                #pyautogui.press('up')
                pyautogui.press('u')
                pyautogui.press('p')
                pyautogui.press('space')
            # down
            if centroid_y >= y1 and centroid_y <=y2:
                print('down')
                #pyautogui.press('down')
                pyautogui.press('d')
                pyautogui.press('o')
                pyautogui.press('w')
                pyautogui.press('n')
                pyautogui.press('space')

                # left-right move
        if centroid_y >= 0 and centroid_y <= y2:
            # left
            if centroid_x >= 0 and centroid_x <= x1:
                print('left')
                #pyautogui.press('left')
                pyautogui.press('l')
                pyautogui.press('e')
                pyautogui.press('f')
                pyautogui.press('t')
                pyautogui.press('space')
            # right
            if centroid_x >= x2:
                print('right')
                pyautogui.press('right')
                pyautogui.press('r')
                pyautogui.press('i')
                pyautogui.press('g')
                pyautogui.press('h')
                pyautogui.press('t')
                pyautogui.press('space')

    cv2.imshow('Original', img)
    k = cv2.waitKey(25)
    if k == 27:
        break

cap.release()
cv2.destroyAllWindows()q
```
